### Para avaliarmos o comportamento da variável RENDA vamos construir uma tabela de frequências considerando as seguintes classes em salários mínimos (SM)
#### <font color='blue'>Descreva os pontos mais relevantes que você observa na tabela e no gráfico.</font>

Classes de renda:

<b>A</b> ► Acima de 25 SM

<b>B</b> ► De 15 a 25 SM

<b>C</b> ► De 5 a 15 SM

<b>D</b> ► De 2 a 5 SM

<b>E</b> ► Até 2 SM

Para construir as classes de renda considere que o salário mínimo na época da pesquisa era de <b>R$ 788,00</b>.

#### Siga os passos abaixo:

### 1º Definir os intevalos das classes em reais (R$)

```python
# Salário mínimo vigente em 2015 — base para definir as classes de renda
salario_minimo = 788

# Pontos de corte das classes, em reais (R$).
# Cada limite = nº de salários mínimos da classe x o salário mínimo.
# A lista começa em 0 e termina na maior renda do dataset (classe A é aberta no topo).
classes = [
    0,                      # início da classe E
    2 * salario_minimo,     # fim de E / início de D  ->  R$ 1.576
    5 * salario_minimo,     # fim de D / início de C  ->  R$ 3.940
    15 * salario_minimo,    # fim de C / início de B  ->  R$ 11.820
    25 * salario_minimo,    # fim de B / início de A  ->  R$ 19.700
    dados['Renda'].max()    # topo da classe A (renda máxima observada)
]

classes
```

### 2º Definir os labels das classes

```python
# Nomes das classes, na MESMA ordem dos bins (renda crescente: E -> A).
# A 1ª posição casa com a 1ª faixa (renda mais baixa) = classe E;
# a última posição casa com a faixa de renda mais alta = classe A.
labels = ['E', 'D', 'C', 'B', 'A']

labels
```

### 3º Construir a coluna de frequências

```python
# pd.cut() encaixa cada renda em uma das 5 classes definidas pelos bins.
# include_lowest=True garante que a renda 0 (limite inferior) entre na classe E.
faixas_renda = pd.cut(
    x = dados['Renda'],
    bins = classes,
    labels = labels,
    include_lowest = True
)

# value_counts() conta quantas pessoas caíram em cada classe
frequencia = faixas_renda.value_counts()

frequencia
```

### 4º Construir a coluna de percentuais

```python
# normalize=True faz o value_counts() devolver proporções (0 a 1) em vez de contagens.
# Multiplicando por 100, obtemos o percentual de pessoas em cada classe.
percentual = faixas_renda.value_counts(normalize = True) * 100

percentual
```

### 5º Juntar as colunas de frequência e percentuais e ordenar as linhas de acordo com os labels das classes

```python
# Junta as duas séries num único DataFrame (alinhadas pelo índice das classes)
tabela_freq_renda = pd.DataFrame(
    {'Frequência': frequencia, 'Porcentagem (%)': percentual}
)

# Ordena de A (maior renda) até E (menor renda) para leitura de cima para baixo
tabela_freq_renda.sort_index(ascending = False)
```

### Construa um gráfico de barras para visualizar as informações da tabela de frequências acima

```python
# Ordena a tabela da classe A (maior renda) até a E (menor renda)
freq_ordenada = tabela_freq_renda.sort_index(ascending = False)

# Define o tamanho da figura (largura, altura em polegadas)
plt.figure(figsize = (8, 5))

# Gráfico de barras: classe de renda no eixo X, percentual de pessoas no eixo Y.
# Cor única e sóbria — a ALTURA da barra já comunica a quantidade; cores
# diferentes por barra só adicionariam ruído visual sem informação nova.
grafico = sns.barplot(
    x = freq_ordenada.index,
    y = freq_ordenada['Porcentagem (%)'],
    color = '#1f6f8b'
)

# Título e rótulos dos eixos (boa prática: nenhum gráfico sem contexto)
grafico.set_title(
    'Distribuição da população por classe de renda — PNAD 2015',
    fontsize = 13, pad = 15
)
grafico.set_xlabel('Classe de renda', fontsize = 11)
grafico.set_ylabel('Percentual de pessoas (%)', fontsize = 11)

# Escreve o valor percentual em cima de cada barra, facilitando a leitura
for container in grafico.containers:
    grafico.bar_label(container, fmt = '%.1f%%', padding = 3)

plt.show()
```

> ### Conclusões

A tabela de frequências e o gráfico de barras revelam uma **concentração extrema de renda na base da pirâmide social brasileira**:

- **Classe E (até 2 salários mínimos): 64,75% das pessoas.** Quase dois em cada três responsáveis por domicílio ganhavam, no máximo, R\$ 1.576,00 com o trabalho principal. Sozinha, esta classe é maior que todas as outras somadas.
- **Classes E e D juntas: 88,96%.** Praticamente nove em cada dez pessoas ganhavam até 5 salários mínimos (R\$ 3.940,00).
- **Classes A e B juntas: apenas 1,62%.** O topo da distribuição — quem ganha acima de 15 salários mínimos — é estatisticamente minúsculo: 1.242 pessoas de um total de 76.840.

O gráfico não desenha uma "escada" suave de desigualdade, mas um **degrau abrupto**: uma barra dominante na classe E e as demais progressivamente menores, sendo A e B quase invisíveis.

Esse resultado **confirma e explica** o que a exploração inicial (`describe()`) já sugeria: a média de renda (R\$ 2.000) era muito maior que a mediana (R\$ 1.200). A tabela mostra a causa — a maioria esmagadora se concentra nas faixas mais baixas, enquanto um grupo minúsculo de altíssima renda (classe A, com valores que chegam a R\$ 200.000) puxa a média para cima, distorcendo-a como retrato do brasileiro "típico". Nesse cenário, **a mediana descreve a realidade melhor que a média**.

**Ressalvas para uma leitura honesta dos dados:**

- O dataset considera apenas as **pessoas de referência** de cada domicílio (o responsável), não toda a população — descreve quem chefia os lares, não cada indivíduo.
- A variável Renda mede apenas o **rendimento do trabalho principal**, sem incluir aposentadorias, aluguéis, benefícios ou outras fontes.
- As classes têm **larguras diferentes** (a classe E cobre uma faixa de 2 SM; a classe C cobre 10 SM). Ainda assim, a concentração na classe E é tão acentuada que a conclusão sobre a desigualdade permanece sólida.

## Realize, para a variável RENDA, uma análise descritiva com as ferramentas que aprendemos em nosso treinamento

### Obtenha a média aritimética

```python
# Média aritmética: soma de todas as rendas dividida pelo número de pessoas
media_renda = dados['Renda'].mean()
media_renda
```

### Obtenha a mediana

```python
# Mediana: o valor central quando todas as rendas são colocadas em ordem.
# Metade das pessoas ganha menos que isso, metade ganha mais.
mediana_renda = dados['Renda'].median()
mediana_renda
```

### Obtenha a moda

```python
# Moda: o valor que mais se repete no conjunto de dados.
# mode() devolve uma Series, pois em tese pode haver mais de uma moda.
moda_renda = dados['Renda'].mode()
moda_renda
```

### Obtenha o desvio médio absoluto

```python
# Desvio médio absoluto: a média das distâncias de cada renda até a média.
# OBS: o antigo método .mad() do pandas foi removido na versão 2.0, então
# calculamos a medida diretamente pela sua fórmula, em 3 passos:
#   1) (Renda - média) -> distância (com sinal) de cada valor até a média
#   2) .abs()          -> torna todas as distâncias positivas
#   3) .mean()         -> média dessas distâncias
desvio_medio_absoluto = (dados['Renda'] - media_renda).abs().mean()
desvio_medio_absoluto
```

### Obtenha a variância

```python
# Variância: a média dos QUADRADOS das distâncias até a média.
# Elevar ao quadrado faz os valores muito distantes (outliers) pesarem muito mais.
# Atenção: a unidade fica "reais ao quadrado", difícil de interpretar sozinha.
variancia_renda = dados['Renda'].var()
variancia_renda
```

### Obtenha o desvio-padrão

```python
# Desvio-padrão: a raiz quadrada da variância.
# A raiz "desfaz" o quadrado e devolve a medida para a unidade original (R$),
# tornando a dispersão finalmente interpretável.
desvio_padrao_renda = dados['Renda'].std()
desvio_padrao_renda
```

### Obtenha a média, mediana e valor máximo da variável RENDA segundo SEXO e COR
#### <font color='blue'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mean', 'median', 'max'}</i></font>

```python
# Tabela cruzada que AGREGA a renda (em vez de apenas contar pessoas):
#   values = dados['Renda']            -> a variável a ser resumida
#   aggfunc = ['mean', 'median', 'max'] -> aplica as três funções de uma vez
renda_sexo_cor = pd.crosstab(
    dados['Sexo'],
    dados['Cor'],
    values = dados['Renda'],
    aggfunc = ['mean', 'median', 'max']
)

# Renomeia os códigos numéricos pelos nomes legíveis
renda_sexo_cor.rename(index = sexo, columns = cor, inplace = True)

renda_sexo_cor
```

> ### Conclusões

O cruzamento da renda com Sexo e Cor revela que a desigualdade brasileira não é um fenômeno único, mas a **soma de duas desigualdades que se empilham**.

**1. A dupla desvantagem: raça e gênero se somam.** Observando as medianas dos grupos majoritários (branco, pardo e preto, que reúnem cerca de 99% do dataset), dois efeitos aparecem com clareza:

### Obtenha as medidas de dispersão da variável RENDA segundo SEXO e COR
#### <font color='blue'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mad', 'var', 'std'}</i></font>

```python
# O antigo 'mad' do pandas foi removido na versão 2.0. Definimos então uma
# função própria para o desvio médio absoluto. Ao passá-la para o aggfunc,
# o pandas usa o NOME da função como rótulo da coluna na tabela.
def desvio_medio_absoluto(serie):
    return (serie - serie.mean()).abs().mean()


# Tabela cruzada com as medidas de DISPERSÃO da renda, por Sexo e Cor
dispersao_sexo_cor = pd.crosstab(
    dados['Sexo'],
    dados['Cor'],
    values = dados['Renda'],
    aggfunc = [desvio_medio_absoluto, 'var', 'std']
)

dispersao_sexo_cor.rename(index = sexo, columns = cor, inplace = True)

dispersao_sexo_cor
```

> ### Conclusões

Enquanto a seção anterior comparou o *centro* da renda de cada grupo, as medidas de dispersão revelam algo complementar: o quão **desiguais são as rendas dentro de cada grupo**.

**1. O grupo branco é o mais desigual internamente.** Entre os grupos estatisticamente robustos (branco, pardo e preto), o desvio-padrão do homem branco (R\$ 4.751) é o maior — bem acima do homem pardo (R\$ 2.312) e do homem preto (R\$ 1.936). Isso significa que o grupo branco concentra, ao mesmo tempo, pessoas de renda muito baixa e as maiores rendas do país (o valor máximo de R\$ 200.000 pertence a um homem branco). O grupo branco não tem apenas a maior renda mediana: tem também a maior distância interna entre seus próprios membros.

**2. A dispersão é ainda mais sensível a valores extremos do que a média.** O desvio-padrão da mulher indígena chega a R\$ 11.957 — de longe o maior valor da tabela, num grupo de apenas 101 pessoas. Esse número não indica que as mulheres indígenas tenham rendas genuinamente muito desiguais entre si; ele reflete a mesma observação extrema de R\$ 120.000 já identificada na seção anterior. Como o desvio-padrão eleva as distâncias ao quadrado, um único outlier o distorce com força ainda maior do que distorce a média. É a confirmação de uma regra prática: **medidas de dispersão exigem ainda mais cautela que medidas de centro em grupos pequenos.**

**Ressalvas para uma leitura honesta dos dados:**

- A variância tem unidade em "reais ao quadrado" e serve sobretudo como passo intermediário; o desvio-padrão (na unidade original, R\$) é a medida de dispersão efetivamente interpretável.
- As comparações de dispersão entre grupos só são confiáveis para branco, pardo e preto. Para indígenas e amarelos, o tamanho reduzido da amostra torna qualquer medida de espalhamento muito instável.
- Dispersão alta não é, por si só, "boa" ou "ruim": ela apenas mede heterogeneidade. A leitura depende do contexto — aqui, indica que dentro do grupo branco convivem realidades econômicas muito distantes.
